```python
#select the properties you want to base your analysis on

properties= ['vertical_group','country_group', 'initial_1k_sims']
#properties=['vertical_group']
```

```python
# Keeping only relevant cols based on the propeties selected

def select_prop(df,properties):
    cols=['sim_id','organization_id','activated_month','assigned_month','max_active_SIMs','active_SIMs_divided_by_months']
    cols.extend(properties)
    df=df[cols]
    return df
```

```python
df=select_prop(df,properties)
df
```

```python
# keep only sims that have been assigned in the desired period
df=df[df.assigned_month>'2019-12-01']
```

```python
# create segmentation month by month on selected properties
def make_segmentation(df, properties):
    cols_for_grouping=['assigned_month','activated_month'] 
    cols_for_grouping.extend(properties)
    grouped=df.groupby(cols_for_grouping).count() # authomatically removes NaNs in activated_month
    grouped.reset_index(inplace=True)

    grouped=grouped[grouped.assigned_month>'2019-12-01']
    grouped.reset_index(inplace=True)
    grouped.groupby(properties).sum()
    grouped.drop('index', axis=1)
    return grouped 
```

```python
segmented=make_segmentation(df,properties)
```

```python
# calculate difference (in months) between assigned and activated month
def month_diff(df):
    df['month']=0
    for i in range(len(df)):
        d_i=datetime.datetime.strptime(df.assigned_month[i], '%Y-%m-%d')
        d_f=datetime.datetime.strptime(df.activated_month[i], '%Y-%m-%d')
        r = relativedelta.relativedelta(d_f,d_i)    
        months = (r.years * 12) + r.months 
        df.month[i]=months
    return df
```

```python
segmented=month_diff(segmented)
```

```python
segmented
```

```python
# get total number of sims assigned in each month per each segment

def get_amounts_segment(df, properties):
    base=['month', 'sim_id']
    base.extend(properties)
    df=df[base]

    base.remove('sim_id')
    df=df.groupby(base).sum() #,'country_group','initial_1k_sims'
    df.reset_index(inplace=True)
    return df
```

```python
tot_amounts=get_amounts_segment(segmented,properties)
```

```python
# get number of sims in each month per each segment with potential to be activated

def get_potential_activation(df,properties):
    potential=df.groupby(properties).count()
    potential.reset_index(inplace=True)
    potential=potential.rename(columns={'sim_id':'to_activate'})
    return potential
```

```python
potential=get_potential_activation(df,properties)
```

```python
potential.to_activate.sum()
```

```python
# get number of sims in each month per each segment that have been activated as % of the total number of sims with potential to be activated

def get_pct(tot_amounts, potential):
    final=pd.DataFrame()
    tot_amounts=tot_amounts.merge(potential[['vertical_group','country_group','initial_1k_sims','to_activate']], on=['vertical_group','country_group','initial_1k_sims'])
    tot_amounts['id'] = list(zip(tot_amounts.vertical_group,tot_amounts.country_group, tot_amounts.initial_1k_sims))
    for i in tot_amounts.id.unique(): # for each segment
        temp=tot_amounts[tot_amounts['id']==i]
        temp.reset_index(inplace=True)
        for ix in range(len(temp)):
            temp.to_activate[ix+1]=  temp.to_activate[ix]- temp.sim_id[ix]
        final=pd.concat([final,temp]) # add to final df
    final['pct']=final.sim_id/final.to_activate # calculate activation pct
    return final
```

```python
final=get_pct(tot_amounts, potential)
```

```python
final=final.drop('index', axis=1)
```
