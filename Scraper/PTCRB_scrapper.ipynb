```python
cont=[1,2] #temporary initialization
i=1 #page counter
dev_type=['other','integration', 'module'] #type of device (to be set in the loop)
df=pd.DataFrame() #df initialization
```

```python
while len(cont)>0:
    #set the type of device you want to scrape from the list dev_type
    data=a+str(i)+b+dev_type[1]+c #set data
    r = re.post('https://certify.ptcrb.com/cfcs/controllers/wordpress.cfc', params=params, headers=headers, data=data)
    json=r.json()
    cont=json['data']
    cont=cont['devices']
    temp=pd.json_normalize(json['data'], record_path =['devices'])
    df=df.append(temp) #append the data to the df
    i+=1 #increment page counter
```

```python
# see if the property value is already in the database (HS)

def get_hs_id_from_linkedin(prop, value):
    uri = "https://api.hubapi.com/crm/v3/objects/companies/search"
    headers = {'authorization': key,'content-type': 'application/json'}
    body = {"filterGroups": [ { "filters": [ { "value": value, "propertyName": prop, "operator": "EQ" } ] }]}
    res = re.post(uri, headers=headers, data=json.dumps(body)).json()
    time.sleep(0.2)
    if len(res['results']) > 0:
        return True
    else:
        return False
```

```python
# see if the property value is already in the database (HS)

def get_hs_id_from_prop(prop, value):
    uri = "https://api.hubapi.com/crm/v3/objects/companies/search"
    headers = {'authorization': key,'content-type': 'application/json'}
    body = {"filterGroups": [ { "filters": [ { "value": value, "propertyName": prop, "operator": "EQ" } ] }]}
    res = re.post(uri, headers=headers, data=json.dumps(body)).json()
    time.sleep(0.2)
    if len(res['results']) > 0:
        return True
    else:
        return False
```

```python
# apply function to check if prop value matches HS company name

unique['HS'] = unique.manufacturer.apply(lambda x: get_hs_id_from_prop('name',x))
```
