```
while len(cont)>0:
    #set the type of device you want to scrape from the list dev_type
    data=a+str(i)+b+dev_type[1]+c #set data
    r = requests.post('https://certify.ptcrb.com/cfcs/controllers/wordpress.cfc', params=params, headers=headers, data=data)
    json=r.json()
    cont=json['data']
    cont=cont['devices']
    temp=pd.json_normalize(json['data'], record_path =['devices'])
    #print(temp)
    df=df.append(temp)
    i+=1
```
